## Task 2: Process Multiple Orders (for loop)

```python
orders = [1200, 2500, 800, 1750, 3000]

total_revenue = 0
discount_count = 0

print("Order Amount -> Discount% -> Final Amount")

for order in orders:

    if order >= 2000:
        discount = 15
    elif order >= 1500:
        discount = 10
    elif order >= 1000:
        discount = 7
    else:
        discount = 0

    final_amount = order - (order * discount / 100)

    print(order, "->", str(discount) + "%", "->", final_amount)

    total_revenue += final_amount

    if discount > 0:
        discount_count += 1

print("\nTotal Revenue:", total_revenue)
print("Orders With Discount:", discount_count)
```

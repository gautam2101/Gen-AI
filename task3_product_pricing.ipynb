```python
price_dict = {
    "Laptop": 55000,
    "Mobile": 25000,
    "Headphones": 2000,
    "Keyboard": 1500,
    "Mouse": 800,
    "Monitor": 12000
}
```

```python
price_dict["Tablet"] = 18000
```

```python
price_dict["Mobile"] = 27000
```

```python
product_name = "Smartwatch"

if product_name in price_dict:
    del price_dict[product_name]
else:
    print(product_name, "not found")

print("Updated Price Dictionary:")
print(price_dict)
```

```python
average_price = sum(price_dict.values()) / len(price_dict)

print("Average Price:", average_price)
```

```python
max_product = max(price_dict, key=price_dict.get)
min_product = min(price_dict, key=price_dict.get)
```

```python
print("Highest Price Product:",
      max_product, price_dict[max_product])
```

```python
print("Lowest Price Product:",
      min_product, price_dict[min_product])
```
